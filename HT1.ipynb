Problema 1

```python
picto = cv2.imread("Discover.jpg")
def colores(cod_color,pic):
  alto = pic.shape[0]
  ancho = pic.shape[1]
  pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
  lienzo = np.zeros((alto, ancho, 3))
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = pic[i, j]
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        if cod_color == 1:
          lienzo[i, j] = [blue, 0, 0]
        elif cod_color == 2:
          lienzo[i, j] = [0, green, 0]
        elif cod_color == 3:
          lienzo[i, j] = [0, 0, red]
        elif cod_color == 10:
          lienzo[i, j] = [0, green, red]
        elif cod_color == 20:
          lienzo[i, j] = [blue, green, 0]
        elif cod_color == 30:
          lienzo[i, j] = [blue, 0, red]

  cv2.imwrite("Problema1.jpg", lienzo)
  Problema1 = cv2.imread("Problema1.jpg")
  Problema1 = cv2.cvtColor(Problema1, cv2.COLOR_BGR2RGB)
  plt.imshow(Problema1) 
  return plt.show()
```

```python
colores(10,picto)
```

**El modelo de color HSV**

El modelo de color HSV se basa en un sistema de color en el que el espacio de color se representa mediante un nico cono. Los tres componentes del cono son tono, saturacin y valor. En el modelo de color HSV, se cumple lo siguiente:

** Tono:** especifica el tono (color) establecido para el color. El tono se indica como un entero del 0 al 240, ambos incluidos. Este se remuestrea a partir de los valores de 0 a 360 en que se indica el tono como ngulo en sentido antihorario alrededor del cono de color. Los colores primarios y secundarios presentan los siguientes valores de tono: rojo = 0 (0), amarillo = 40 (60), verde = 80 (120), cian = 120 (180), azul = 160 (240) y magenta = 201 (300).

** Saturacin:** especifica la intensidad de saturacin establecida para el color. La saturacin se indica como un entero del 0 al 255, ambos incluidos (que representan del 0 al 100%). La saturacin de un color hace referencia a la medida en que se aleja de un color neutro como el gris o, dicho de un modo ms sencillo, su colorido. Cuando la saturacin es de 255, el color presenta una saturacin mxima. Cuando el valor de la saturacin es 0, el color es no saturado y parece gris (a menos que el valor sea 0  255, en cuyo caso aparece negro o blanco).

** Brillo:** especifica la intensidad del blanco en el color. El valor se indica como un entero del 0 al 255, ambos incluidos (que representan del 0 al 100%). Un color con un valor establecido en 0 aparece negro. Un color con un valor establecido en 255 y una saturacin de 0 aparece como blanco. Conversin al espacio de colores HSV En lugar de eso vamos a convertir los pxeles desde el espacio RGB al espacio HSV, en donde cada pxel se codifica tambin con tres valores, pero ahora con significado diferente: El canal H indica el tono (hue) del color (verde, rojo, azul, amarillo, etc). El tono es lo que tradicionalmente entendemos como "color". El canal S indica la saturacin (saturation) de color Si un color es ms saturado, es ms "puro" Si es menos saturado, tiene ms "gris" El canal V indica la luminosidad o intensidad (value) del pixel. V=0 indica un pixel negro V=1 indica un pixel blanco La funcin rgb_to_hsv del mdulo matplotlib.colorsnos permite hacer justamente eso.
